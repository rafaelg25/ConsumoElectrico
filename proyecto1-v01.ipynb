```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
fig, axes=plt.subplots(1,3)
fig.set_size_inches(15,5)
axes[0].plot(data.index,y)
plot_acf(y,lags=48,ax=axes[1]);
plot_pacf(y,lags=48,ax=axes[2]);
#La autocorrelaci贸n es de la misma serie. (tiene un proceso sistematico) con un lags de 48 (se desplaza en una venta) es estadisticamente significativa
#la autocorrelaci贸n parcial el desplazamiento relevante es el 2
#para los datos de CNAVIA
```

```python
fig, axes=plt.subplots(1,3)
fig.set_size_inches(15,5)
axes[0].plot(data1.index,y2)
plot_acf(y2,lags=48,ax=axes[1]);
plot_pacf(y2,lags=48,ax=axes[2]);
#La autocorrelaci贸n es de la misma serie. (tiene un proceso sistematico) con un lags de 48 (se desplaza en una venta) es estadisticamente significativa
#la autocorrelaci贸n parcial el desplazamiento relevante es el 3
#para los datos de AJAHUEL
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(y,maxlag=1)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
#el conjunto de datos es estacionario para la subestaci贸n CNAVIA
#donde p<0,05
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python

modelo=ARIMA(y_train,order=(2,0,3))
predictor=modelo.fit()
print("MSE",predictor.mse)
predictor.summary()
#para sub estaci贸n CNAVIA
#p de la parte autocorrelaci贸n 2 unidad
#q de la parte autocorrelacion parcial 3 unidades (media movil)
```

```python
modelo1=ARIMA(y_train1,order=(2,0,4))
predictor1=modelo1.fit()
print("MSE",predictor1.mse)
predictor1.summary()
#para sub estaci贸n AJAHUEL
#p de la parte autocorrelaci贸n 2 unidad
#q de la parte autocorrelacion parcial 4 unidades (media movil)
#para este serie se ofrece un mejor BIC
```

```python
#la prediccion de CNAVIA
modelo_pred=predictor.get_forecast(steps=len(y))
modelo_ci=modelo_pred.conf_int(0.05)
```

```python
plt.plot(data.index,y,modelo_pred.predicted_mean)
plt.fill_between(y,modelo_ci[:,0],modelo_ci[:,1],color="b",alpha=.15)
#esto nos comprueba que el modelo contien mucho error
```
